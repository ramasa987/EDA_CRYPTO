## 📈💰📊 Definicion de las columnas:

**top_500_metadata.csv:** Contiene datos de las 500 monedas más populares.

-**id:** Identificador único de CoinGecko (por ejemplo, bitcoin, solana).

-**symbol:** Símbolo bursátil (por ejemplo, btc, sol).

-**name:** Nombre completo del activo.

-**image:** URL del logo de la moneda.

-**current_price:** Último precio de mercado en USD.

-**market_cap:** Capitalización bursátil total en USD.

-**market_cap_rank:** Clasificación global por capitalización bursátil.

-**fully_diluted_valuation:** Capitalización de mercado teórica si la máxima oferta estuviera en circulación.

-**total_volume:** Volumen de negociación las 24 horas.

-**high_24h / low_24h:** Precio más alto y más bajo en las últimas 24 horas.

-**price_change_24h:** Cambio absoluto de precio en las últimas 24 horas.

-**price_change_percentage_24h:** Porcentaje de cambio en el precio en las últimas 24 horas.

-**circulating_supply:** Cantidad de monedas actualmente en el mercado.

-**total_supply:** Cantidad total de monedas creadas.

-**max_supply:** Cantidad máxima de monedas que existirá jamás.

-**ath:** Precio máximo histórico.

-**atl:** Precio más bajo de todos los tiempos.

-**last_updated:** Marca temporal de la instantánea de metadatos.
