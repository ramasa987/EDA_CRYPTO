## 🌟 Descripción del Proyecto
Este proyecto presenta un análisis exploratorio de datos (EDA) la rentabilidad de las 10 criptomonedas mas importates seleccionadas a tra ves de unos criterios establecidos como:
- **Comparación de Capitalización de mercado (Market Cap)**
- **Análisis de Volatilidad en 24 Horas**
- **Análisis de Volumen de Trading**
- **Análisis de Máximos Históricos (ATH)**


---

## 💡 Hipótesis del proyecto:

- **¿Qué Criptomoneda tuvo el valor de cotización más alto del año 2025?**
- **¿En qué mes se registraron las mayores caídas y en cual los mayores incrementos?**
- **¿Cuál ha sido la criptomoneda mas más rentable en este año? (utilizamos el ROI)**
- **Average change (Volatilidad de la inversión)**
